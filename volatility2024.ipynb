```python
import pandas as pd
import helpermodules.memory_handling as mh
import numpy as np
```

```python
# Load the preprocessed data
data = pd.read_csv("sp500minute.csv")
speeches = pd.read_csv("fedspeeches.csv")
```

```python
speeches['date'] = pd.to_datetime(speeches['date'])
filtered_speeches = speeches[speeches['date'] >= pd.to_datetime('2024-01-01')]
data['Datetime'] = pd.to_datetime(data['<Date>'] + ' ' + data[' <Time>'],format='%d/%m/%Y %H:%M:%S')
# Drop the <Date> and <Time> columns
data = data.drop(columns=['<Date>', ' <Time>'])

# Reorder columns to place Datetime as the first column
cols = ['Datetime'] + [col for col in data.columns if col != 'Datetime']
data = data[cols]

# Display the updated DataFrame
```

```python
filtered_speeches['date'] = pd.to_datetime(filtered_speeches['date'])

# Set the start time to 10:00 AM EST for each speech
filtered_speeches['Start Time'] = filtered_speeches['date'] + pd.Timedelta(hours=10)
```

```python

filtered_speeches["Duration"] = filtered_speeches['text'].apply(lambda x: max(1, len(x.split()) / 130))
filtered_speeches
```

```python
filtered_speeches["Duration"] = filtered_speeches["Duration"].round()
filtered_speeches
```

```python
filtered_speeches["End time"] = filtered_speeches["Start Time"] + pd.to_timedelta(filtered_speeches["Duration"], unit='m')
filtered_speeches
```

```python
# Assuming 'df' is your DataFrame
column_names = list(data.columns)
print(column_names)
```

```python
sp500 = data
# Define a function to calculate volatility
def calculate_volatility(prices):
    returns = np.log(prices / prices.shift(1))
    return np.std(returns)

# Initialize a list to store results
volatility_results = []

# Loop through each speech
for _, speech in filtered_speeches.iterrows():
    # Extract the corresponding time period from S&P 500 data
    speech_period_data = sp500[
        (sp500['Datetime'] >= speech['Start Time']) &
        (sp500['Datetime'] <= speech['End Time'])
    ]

    # Ensure there is enough data to calculate volatility
    if len(speech_period_data) > 1:
        volatility = calculate_volatility(speech_period_data[' <Close>'])
        volatility_results.append({
            'speaker': speech['speaker'],
            'title': speech['title'],
            'Start Time': speech['Start Time'],
            'End Time': speech['End Time'],
            'volatility': volatility
        })

# Convert results to a DataFrame
volatility_df = pd.DataFrame(volatility_results)

# Rank speeches by volatility
volatility_df = volatility_df.sort_values('volatility', ascending=False)

# Print or save the speeches with the highest volatility
print(volatility_df.head())

# Save to a CSV file for further analysis
volatility_df.to_csv("speech_volatility_analysis.csv", index=False)
```
